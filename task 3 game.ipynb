```python
def check_digit():
    digit=str(input("please enter your character"))
    length=len(digit)
    if length==1:
        return digit
    else:
        check_digit()
        
        
```

```python
import random as rn

player_name=str(input("please enter your name >>"))
a=['red','yellow','blue','gold','green','browen']
word=rn.choice(a)
#print(word)
le=len(word)

for c in word:
    print("*")
erorr=7
list1=[]
list2=[]

while True:
    
    d=check_digit()
    list1.append(d)
    if d not in word:
        erorr-=1
        print(f"the rest of you {erorr} attempts")
        
        
    for i in word:
        
        if i in list1:
            print(i)
            list2.append(i)
            
        else:
            print("*")
    
    if erorr<=0:
        break
    s=0
    ss=0
    set1=set(list2)
    for x in word:
        if x in set1:
            s+=1
        else:
            ss=1
    if ss==0:
        break
   

        
if erorr>0 and ss==0:
    print(f'{player_name} is win')
    
elif len(list1)<le or erorr<=0  :
    print("good luck")
   
        
        
        
    
    
    
    
    
```
